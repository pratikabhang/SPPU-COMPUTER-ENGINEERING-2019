```python
df=df.replace(to_replace=(0,1), value=('D','A'))
df
```

```python
df['Survived']=df['Survived'].replace(to_replace=(0,1), value=('D','A'))
df
```

```python
dummies=pd.get_dummies(df.Survived)
A=pd.concat([df,dummies],axis=1)
A
```

```python
dummies=pd.get_dummies(df,columns=['Survived'],dtype=int)
A=pd.concat([df,dummies],axis=1)
A
```

```python
df['Fare'] = pd.to_numeric(df['Fare'], errors='coerce')
df['Fare'].dropna(inplace=True)
sns.displot(df['Fare'], kde=False)
```

```python
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['Age'].dropna(inplace=True)
sns.displot(df['Age'], kde=False)
```

```python
df['Pclass'] = pd.to_numeric(df['Pclass'], errors='coerce')
df['Pclass'].dropna(inplace=True)
sns.displot(df['Pclass'], kde=False)
```
