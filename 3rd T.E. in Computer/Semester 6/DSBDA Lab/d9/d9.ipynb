```python
df['age']=df['age'].replace(np.nan,df[ 'age'].mean()) 
```

```python
#Distribution Plots
df = sns.load_dataset('titanic')
sns.distplot(x=df['age'])
```

```python
#Joint Plot
df = sns.load_dataset('titanic')
sns.jointplot(x=df['age'], y=df['fare'], kind='scatter')
```

```python
#Bar Plot
sns.barplot(x='sex', y='age', data=df, estimator=np.std)
```

```python
#Count Plot
sns.countplot(x='sex', data=df) 
```

```python
#Box Plot
sns.boxplot(x='sex', y='age', data=df)
```

```python
#Violin Plot
sns.violinplot(x='sex',y='age',data=df) 
```

```python
#Strip Plot
sns.stripplot(x='sex', y='age', data=df)
```

```python
# Plot a histogram of the ticket prices
dataset = sns.load_dataset('titanic')
sns.histplot(dataset['fare'], kde=False, bins=10)
```
